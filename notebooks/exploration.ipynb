# M3Benchmark Data Exploration

This notebook provides a starting point for exploring the M3Benchmark dataset.

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

## Load Dataset

```python
# List available data files
data_dir = Path('../train_chunked')
data_files = list(data_dir.glob('*.json'))
print(f"Found {len(data_files)} data files:")
for f in data_files[:10]:
    print(f"  - {f.name}")
```

```python
# Load a sample file
sample_file = data_files[0] if data_files else None
if sample_file:
    with open(sample_file, 'r') as f:
        data = json.load(f)
    print(f"Loaded: {sample_file.name}")
    print(f"Type: {type(data)}")
    print(f"Length: {len(data) if isinstance(data, list) else 'N/A'}")
```

## Explore Data Structure

```python
# Examine first sample
if data and isinstance(data, list):
    sample = data[0]
    print("Keys:", list(sample.keys()))
    print("\nSample:")
    for key, value in sample.items():
        print(f"  {key}: {type(value).__name__}")
```

```python
# Pretty print a sample
if data and isinstance(data, list):
    print(json.dumps(data[0], indent=2)[:2000])
```
